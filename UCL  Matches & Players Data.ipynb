### In order to make a holistic evaluation I am going to assign weight numbers to the each attribute in these dataframes below. With this approach I am going to evaluate all the players in these datasets from 0 point to 100. I am going to also explain why I gave which weight number to the which attribute in the comments.

```python
#df_attack
#df_attempts
#df_defending
#df_disciplin
#df_distribution
#df_goalkeep
```

```python
#I'm starting with Goalkeeper
df_goalkeep.head()
```

### Comment:

```python
#I have the highest and the lowest weight numbers to 'saved' and 'conceded' columns.
#Because for a goalkeeper these are the most important two aspects. Saving goals means you are contributing for a possible win.
#Whereas conceding goals means possible loss.
#'saved_penalties' is the second highest weight number (0.4) point because compared to the rest saving penalty is the most important.
#'clean_sheets' is 0.3 point because it shows persistence.
#'punches_made' is 0.2 point because punching is an effective move for a goalkeeper to deflect crosses.
#I also gave point to the 'match_played' column beacuse being with your team in every match is a dedication.
#Also means the player is important for the team.
df_gk_score=df_goalkeep['performance_score_gk'] = (
    0.6 * df_goalkeep['saved'] +
    (-0.6) * df_goalkeep['conceded'] +
    0.4 * df_goalkeep['saved_penalties'] +
    0.3 * df_goalkeep['cleansheets'] +
    0.2 * df_goalkeep['punches made'] + 
    0.1 * df_goalkeep['match_played']
)
```

```python
df_gk_scores = df_goalkeep[['player_name','position','club','match_played','conceded','performance_score_gk']].copy()
df_gk_scores.reset_index(drop=True, inplace=True)
df_gk_scores.head()
```

```python
#Calculating the point of the player on 100.
#player_score_100 = (player_performance_score - min_performance_score) / (max_performance_score - min_performance_score) * 100
```

```python
min_performance = df_gk_scores['performance_score_gk'].min()
max_performance = df_gk_scores['performance_score_gk'].max()

# Scale the performance to a range from 0 to 100
df_gk_scores['performance_gk_100'] = ((df_gk_scores['performance_score_gk'] - min_performance) / (max_performance - min_performance)) * 100
```

### Performance scores of all the goalkeepers are now evaluated from 0 to 100. In the "df_gk_scores" dataframe.

```python
df_gk_scores.head()
```

### Now I am going to do the same operation for each footballer traits :Attempt,Defending,Disciplin,Distribution

```python
#Resuming with defence position
df_defending.head()
```

### Comment:

```python
#'balls_recovered' (0.6): 
#If the ball is in your control,your opponent can't score.One of the fundamental aspects of defence.
#Therefore I gave 0.6 point to the 'balls recovered'

#'t_won'/'t_lost' +0.4/-0.4 respectively:
#Both of these defending traits have the same importance.
#Winning tackles by a defence player could lead so serious goal opportunities
#However loosing tackle,giving the control of the ball to the opponent has the equal opportunity
#(Especially if the defence player who lost tackle is the last player of the defence line)

#'clearance_attempted' 0.3:
#Getting the ball away from a dangerous zone is of course an important skill.
#But even though you send the ball away, opponent can recover the ball quickly.
#Because in clearance there is no immediate target regarding a recipient for the ball.


#'match_played' 0.1:
#Being with your team in every match is a dedication.
#Also means the player is important for the team.


df_defending_score=df_defending['performance_score_def'] = (
    0.6 * df_defending['balls_recoverd'] +
    0.4 * df_defending['t_won'] +
    (-0.4) * df_defending['t_lost'] +
    0.3 * df_defending['clearance_attempted'] +
    0.1 * df_defending['match_played'] 
)
```

```python
df_defending_score.head()
```

```python
df_defending_score = df_defending[['player_name', 'position','club', 'match_played','performance_score_def']].copy()
df_defending_score.reset_index(drop=True, inplace=True)
df_defending_score
```

```python
min_performance = df_defending_score['performance_score_def'].min()
max_performance = df_defending_score['performance_score_def'].max()

# Scale the performance to a range from 0 to 100
df_defending_score['performance_def_100'] = ((df_defending_score['performance_score_def'] - min_performance) / (max_performance - min_performance)) * 100
```

### Each player's defence score is now evaluated with a score from 0 to 100.

```python
df_defending_score.head()
```

### Attempt:

```python
#Now starting to attempts dataframe
df_attempts.head()
```

### Comment:

```python
#'on_target' (0.9):
#Because shots that on target are extremely valuable for a possible goal.
#Especially if the team's offensive players are good at chasing the position.

#'off_target' (-0.4):
#'blocked' (-0.5):
#Both of these are basically wasting a possible goal.

df_attempts_score=df_attempts['performance_score_attempt'] = (
    0.9 * df_attempts['on_target'] +
    (-0.4) * df_attempts['off_target'] +
    (-0.5) * df_attempts['blocked'] +
    0.1 * df_attempts['match_played'] 
)   
```

### Comment:

```python
#for "df_disciplin" dataset "fouls_committed" attribute's weight number should be negative at first glance.
#However there are some moments during the match that a player commits foul intentionally, in order to prevent bigger danger.
#For example we can see a defence player pulls the jersey or arm of his/her rival, if the rival is going to be 1 to 1 with the goalkeeper.
#But this aspect of committing foul is not mentioned in the dataset.
#Therefore I will treat "fouls_committed" attribute of this dataset as negative.
```

### 1) Make the best 11 based on their performance scores

```python
#I'm going to write a fucntion to search and find the top 2 players according to their performance.
def find_top_players(df, column_name, n=2):
    # Assuming the DataFrame has a 'player_name' column
    top_players = df.nlargest(n, column_name)['player_name'].values
    return top_players
# Example usage:
top_discipline_players = find_top_players(df_disciplin_score, 'performance_disciplin_100', n=2)
top_attack_players = find_top_players(df_attack_score, 'performance_attack_100', n=2)
top_attempts_players = find_top_players(df_attempts_score, 'performance_attempt_100', n=2)
top_distribution_players = find_top_players(df_distribution_score, 'performance_dist_100', n=2)
top_defending_players = find_top_players(df_defending_score, 'performance_def_100', n=2)
top_gk_players = find_top_players(df_gk_scores, 'performance_gk_100', n=2)

print("Top Discipline Players:", top_discipline_players)
print("Top Attack Players:", top_attack_players)
print("Top Attempts Players:", top_attempts_players)
print("Top Distribution Players:", top_distribution_players)
print("Top Defending Players:", top_defending_players)
print("Top Goalkeeper Players:", top_gk_players)
```

### 3)Which team scored the most and which team conceded the least?

```python
#First I'll  find the total goals scored by each team
total_goals=df_key_stats.groupby('club')['goals'].sum().reset_index()
total_goals.head()
```

```python
#Then I am going to extract the total goals conceded from the df_goalkeep dataframe
total_concede=df_goalkeep.groupby('club')['conceded'].sum().reset_index()
total_concede.head()
```

```python
#Then I am going to merge scored goals and conceded goals dataframe 
total_goals = total_goals.merge(total_concede[['club', 'conceded']], on='club', how='left')
total_goals.head()
```

```python
#Now I'd like to add an "average" column to see difference between scored and conceded goals.
total_goals['average']=(total_goals['goals']-total_goals['conceded'])
total_goals.head()
```

```python
#Sorting the output
total_goals.sort_values(by='average',ascending=False)
```

### Finally visualizing the dataframe to have a better understanding about team's scored vs conceded goals.

```python
plt.figure(figsize=(12, 6))
plt.bar(total_goals['club'], total_goals['average'])
plt.xlabel('Teams')
plt.ylabel('Average')
plt.title("Team's Success According to Average")
plt.xticks(rotation=90)  
# Show the plot
plt.tight_layout()
plt.show()
```
